```python
import pandas as pd
url = 'https://raw.githubusercontent.com/sabah0312/ml-datasets/refs/heads/main/regression_toy_data'
df = pd.read_csv(url,   delim_whitespace=True)

df
```

```python
import torch

X = torch.tensor(df['X'].values, dtype= torch.float32)
y = torch.tensor(df['y'].values, dtype= torch.float32)
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
class Perceptron:
  def __init__(self, num_features, learning_rate = 0.01):
    self.linear = torch.nn.Linear(num_features, 1)
    self.learning_rate = learning_rate
    self.optimizer = torch.optim.SGD(self.linear.parameters(), lr = learning_rate)
    self.loss_fn = torch.nn.MSELoss()

  def forward(self, x):
    return  self.linear(x)

  def update(self, x, y_true):
    predictions = self.forward(x)
    loss = self.loss_fn(predictions, y_true)
    loss.backward()
    self.optimizer.step()
    self.optimizer.zero_grad()
    return loss
```

```python
PPn = Perceptron(1)

print(f"initial weights: {PPn.linear.weight}")
print(f"initial bias : {PPn.linear.bias}")
epoch_no = 150
for epoch in range(epoch_no):
  loss = PPn.update(X, y)
  if epoch%10 == 0:
    print(f"epoch: {epoch}/ {epoch_no}, loss: {loss}")
print(f" weights: {PPn.linear.weight}")
print(f" bias : {PPn.linear.bias}")
```

```python
class LR(torch.nn.Module):
  def __init__(self, num_features, learning_rate = 0.01):
    super().__init__()
    self.linear = torch.nn.Linear(num_features, 1)
    self.learning_rate = learning_rate
    self.optimizer = torch.optim.SGD(self.linear.parameters(), lr = learning_rate)
    self.loss_fn = torch.nn.MSELoss()

  def forward(self, x):
    return  self.linear(x)

  def update(self, x, y_true):
    predictions = self.forward(x)

    loss = self.loss_fn(predictions, y_true)

    loss.backward()
    self.optimizer.step()
    self.optimizer.zero_grad()

    return loss
```

```python
PPn = Perceptron(1)

print(f"initial weights: {PPn.linear.weight}")
print(f"initial bias : {PPn.linear.bias}")
epoch_no = 150
for epoch in range(epoch_no):
  loss = PPn.update(X, y)
  if epoch%10 == 0:
    print(f"epoch: {epoch}/ {epoch_no}, loss: {loss}")
print(f" weights: {PPn.linear.weight}")
print(f" bias : {PPn.linear.bias}")
```
